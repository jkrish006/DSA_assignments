```python
# Generate a 5×4 array with random values between 0 and 1 (float).
# ● Find the row with the maximum sum of values.
# ● Normalize each row using the formula:
matrix2=np.random.rand(5,4)
matrix2
```

```python
row_sums = np.sum(matrix2, axis=1) #adding across rows
row_sums
```

```python
max_row=np.argmax(row_sums)
max_row
```

```python
print(matrix2[max_row])
```

```python
normalized = (matrix2 - matrix2.min(axis=1, keepdims=True)) / \
             (matrix2.max(axis=1, keepdims=True) - matrix2.min(axis=1, keepdims=True))
normalized
```
