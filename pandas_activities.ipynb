```python
import numpy as np
import pandas as pd
```

```python
#  Load the Titanic dataset and display the first 10 rows.
data=pd.read_csv("/content/titanic_dataset.csv")
data
```

```python
data.head(10)
```

```python
# Find the number of missing values in each column.
data.isnull().sum()
```

```python
# Find the percentage of passengers who survived, grouped by gender.
data.groupby("Sex")["Survived"].mean()*100
```

```python
data["Age"].describe()
```

```python
# Find the average, median, and standard deviation of passenger ages
data['Age'].agg(['mean', 'median', 'std'])
```

```python
# Count how many passengers were in each class (pclass).
data["Pclass"].value_counts()
```

```python
# Show all passengers who were female, under age 18, and survived.
result= data[
    (data["Sex"]== "female")&
    (data["Age"]<18)&
    (data["Survived"]==1)
]
result
```

```python
# Replace missing age values with the median age.
data["Age"].median()

```

```python
data["Age"] = data["Age"].fillna(28)
data
```

```python
# Create a column age_group with values:
# a. 'Child' for age < 18
# b. 'Adult' for age 18–60
# c. 'Senior' for age > 60

data["Age_group"] = data["Age"].apply(
    lambda Age:"Child" if Age < 18 else "Adult" if Age<=60 else "Senior"
    )
data["Age_group"]
```

```python
#  Display the top 5 passengers who paid the highest fare.

data.loc[data["Fare"].idxmax()]
```

```python
#  Find the average fare paid, grouped by both pclass and sex.
data.groupby(["Pclass","Sex"])["Fare"].mean()
```

```python
#  Rename the following columns:
# a. 'sex' → 'gender'
# b. 'fare' → 'ticket_fare'
data=data.rename(columns={"Sex":"Gender", "Fare":"Ticket_fare"})
data
```

```python
#  Count the number of passengers who survived and didn’t survive, grouped by class.

sort_data=data.groupby(["Pclass","Survived"]).size().unstack()
sort_data.columns=["Not survived","Survived"]
sort_data
```

```python
#  Find the average age of passengers grouped by both class and gender.

data.groupby(["Pclass","Gender"])["Age"].mean().unstack()
```
